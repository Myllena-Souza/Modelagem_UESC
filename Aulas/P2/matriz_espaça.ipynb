Calcular os valores de beta

```python
from IPython.display import Image, display

display(Image(filename='exp1.png', width=600))
```

```python
# CALCULANDO OS VALORES DE BETA

beta1 = c1/b1
beta2 = c2/(b2-a2*beta1)
beta3 = c3/(b3-a3*beta2)
beta4 = c4/(b4-a4*beta3)
beta5 = c5/(b5-a5*beta4)
beta6 = c6/(b6-a6*beta5)
```

```python
print("-------------------------------")
print("RESULTADOS")
print("-------------------------------")
print(" O resultado de beta1 = {:.2f}".format(beta1))
print(" O resultado de beta2 = {:.2f}".format(beta2))
print(" O resultado de beta3 = {:.2f}".format(beta3))
print(" O resultado de beta4 = {:.2f}".format(beta4))
print(" O resultado de beta5 = {:.2f}".format(beta5))
print(" O resultado de beta6 = {:.2f}".format(beta6))
print("-------------------------------")
```

### CRESOLVENDO USANDO BIBLIOTECAS PYTHON.

```python
# DADOS DE ENTRADA
# DECLARANDO AS VARIAVEIS DIAGONAIS EM FORMA DE LISTA
# SÃO 7 LINHAS A MATRIZ

#diagonal a é a diagonal inferior
a = [1, 1, 1, 1, 1, 1, 1]

#diagonal b é a diagonal principal
b = [-2, -2, -2, -2, -2, -2, -2]

#diagonal c é a diagonal superior
c = [1, 1, 1, 1, 1, 1, 1]

# declarando o vetor d
d = [-240, -40, -40, -40, -40, -40, -60]
```

```python
# declarando os valores de beta.
beta = [0]*6

# declarando os valores de beta.
gama = [0]*7

# declaração dos resultados
x =[0]*7
z =[0]*7
```

```python
# EXPRESSÕES DE CALCULO

# DETERMINANDO BETA (utiliza laço for)

for i in range(0, 6):
    if i == 0:
        beta[i] = c[i]/b[i]
    else:
        beta[i] = c[i]/(b[i] - a[i]*beta[i-1])
```

```python
# DETERMINANDO GAMA (utiliza laço for)

for i in range(0, 7):
    if i == 0:
        gama[i] = d[i]/b[i]
    else:
        gama[i] = (d[i]-a[i]*gama[i-1]) / (b[i]-a[i]*beta[i-1])
```

```python
# DETERMINAÇÃO DE X (xi = gamai - betai*xi+1)

for i in range(6, -1, -1):
    if i == 6:
        x[i] = gama[i]
    else:
        x[i] = gama[i] - beta[i]*x[i+1]
```

```python
# CONFERINDO OS RESULTADOS

for i in range(0, 7):
    if i == 0:
        z[i] = -2*x[i] + x[i+1]
    elif i>0 and i<6:
        z[i] = x[i-1] -2*x[i] + x[i+1]
    else:
        z[i] = x[i-1] -2*x[i]
```

```python
# IMPRIMINDO RESULTADOS


print("Os valores de beta são: ")
print("---------------------------")
for i in range(0, 6):
    print("O valor de beta{} é = {:.2f}".format(i+1, beta[i]))
print("---------------------------")

print("Os valores de gama são:")
print("---------------------------")
for i in range(0, 7):
    print("O valor de gama {} é = {:.2f}".format(i+1, gama[i]))
print("---------------------------")
print("Os valores de x são:")
print("---------------------------")
for i in range(7, 0, -1):
    print("O valor de x {} é = {:.2f}".format(i, x[i-1]))
print("---------------------------")
print("Os valores de z são:")
print("---------------------------")
for i in range(0, 7):
    print("O valor de z {} é = {:.2f}".format(i+1, z[i]))
print("---------------------------")
```
