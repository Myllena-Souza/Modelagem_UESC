**FATOR DE SEPARAÇÃO:**

#### Fator de Separação

O **Fator de Separação** é definido como a relação entre a $\underline{vazão \ molar \ que \ sai}$ no
topo sobre a $\underline{vazão  \ molar \ que  \ entra}$ no separador de um dado componente.



\begin{equation}
   α_{1} =  n_{51}/n_{31}
\end{equation}


\begin{equation}
   α_{2} =  n_{52}/n_{32}
\end{equation}

\begin{equation}
   α_{3} =  n_{53}/n_{33}
\end{equation}
