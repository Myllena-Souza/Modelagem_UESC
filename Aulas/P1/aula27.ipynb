### PASSO 1:

Vamos inicialmente cálcular o determinante da matriz $A$ - $det(A)$.

**Solução**: 

Vamos resolver a matriz: 

$$ A.x=b $$

para determinar as incógnitas $x$ de forma que:

$$ x = A^{-1}.b $$

Sendo $ A^{-1} $ a matriz inversa.

O cálculo da matriz inversa pode ser obtido por:

$$ A^{-1} = \frac{1}{det(A)},Adj(Ac) $$

Sendo:  
* $ det(A) $ o determinante da matriz $ A $, e 

* $ Adj(Ac) $ a matriz adjunta de $ A $ ou matriz transposta $ A^{T} $.

A matriz cofator $ A_c $ é dada por $ A_c = [c_{ij}] $, sendo $ c_{ij}= (-1)^{i+j}a_{ij} $

__Calculo da inversa da matriz $ A_ij $.__

A inversa da matriz cofator $ A^{T} $ (matriz transposta) é feita transformando linhas em colunas da matriz. Vamos fazer essa transformação calculando algebricamente, mas depois vamos calcular utilizando Numpy.

```python
AT = np.array ([[c11, c12, c13], [c21, c22, c23], [c31, c32, c33]])
AT
```

__Calculo da matriz inversa  $ A^{-1} $.__

Vamos aplicar a expressão abaixo para calcular a matriz inversa.

$$ A^{-1} = \frac{1}{det(A)}.A^{T} $$

Sendo:  
* $ det(A) $ o determinante da matriz $ A $, e 

* $ A^{T} $ a matriz transposta.


```python
# As linhas viram colunas
A_inv = np.array([
    [c11/d, c21/d, c31/d], [c12/d, c22/d, c32/d], [c13/d, c23/d, c33/d] 
])
A_inv
```

__Calculo do valor de b__

```python
b = np.array([30.0, 25.0, 10.0])
b
```

```python
# calculo de x = A^-1.b usando método algebrico.
# Extraindo os valores de b, armazenados em cada celula
b1 = b[0]
b2 = b[1]
b3 = b[2]

m1 = A_inv[0][0]*b1 + A_inv[0][1]*b2 + A_inv[0][2]*b3
x1 = print("O valor de m1 = {:.2f}".format(m1), "Kg/s")
```

```python
m2 = A_inv[1][0]*b1 + A_inv[1][1]*b2 + A_inv[1][2]*b3
x2 = print("O valor de m2 = {:.2f}".format(m2), "Kg/s")
```

```python
m3 = A_inv[2][0]*b1 + A_inv[2][1]*b2 + A_inv[2][2]*b3
x3 = print ("O valor de m3 = {:.2f}".format(m3), "Kg/s")
```

```python
import numpy as np
import numpy.linalg as nl
```

```python
# Vamos criar a matriz A. 
A = np.array([
    [a11, a12, a13],
    [a21, a22, a23],
    [a31, a32, a33]
    ])

Inversa_A = np.linalg.inv(A)
# vamos criar a matris do lado direito - b
b = np.array([30.0, 25.0, 10.0])

# VAmos resolver o sistema linear usando np.solve()
x = nl.solve(A, b)
m1 = x[0]
m2 = x[1]
m3 = x[2]
```
